```python
!pip install Sastrawi
import csv
import pandas as pd
import numpy
from sklearn.feature_extraction.text import CountVectorizer
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
url = open ("/content/dataset_tweet_sentiment_opini_film.csv", 'r')
names = ["Id", "Sentiment", "Text Tweet"]
data = pd.read_csv(url, names=names)
data.head()
```

```python
#validasi
word_X = data["Id"]
word_Y = data["Sentiment"]
word_Z = data["Text Tweet"]
```

```python
def hitung_token(wm, feat_names):
    doc_names = ['data{:d}'. format(idx) for idx, _ in enumerate(wm)]
    df=pd.DataFrame(data=wm.toarray(), index=doc_names, columns=feat_names)

    return (df)

token = CountVectorizer(lowercase=True)
wm = token.fit_transform(word_Z)
hasil_token = token.get_feature_names()

hitung_token(wm, hasil_token)
```

```python
# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def hitung_token(wm, feat_names):
    doc_names = ['data{:d}'. format(idx) for idx, _ in enumerate(wm)]
    df=pd.DataFrame(data=wm.toarray(), index=doc_names, columns=feat_names)

    return (df)

token = CountVectorizer(lowercase=True)
wm = token.fit_transform(word_Z)
hasil_token = token.get_feature_names()

hasil_stemming=[]
for kata in hasil_token:
        hasil_stemming.append(stemmer.stem(kata))

hasil_stemming
```
